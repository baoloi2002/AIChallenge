```python
import os


def scan_file_path_in_folder(directory):
    lst = []
    for root, dirs, files in os.walk(directory):
        # check files in the directory
        for file in files:
            lst.append(os.path.join(root, file))
    return lst


def scan_file_name_in_folder(directory):
    lst = []
    for root, dirs, files in os.walk(directory):
        # check files in the directory
        for file in files:
            # remove the extension
            file_name = os.path.splitext(file)[0]
            lst.append(file_name)
    return lst


path = r"Data\rawData\map-keyframes"
print(scan_file_name_in_folder(path))
print(scan_file_path_in_folder(path))
```

```python
import torch
from transformers import pipeline
import pandas as pd
from pydub import AudioSegment
import numpy as np


# Function to convert AudioSegment to a NumPy array


def audiosegment_to_numpy(audio_segment):
    samples = np.array(audio_segment.get_array_of_samples())
    return samples.astype(np.float32) / (2**15)  # Convert to float32 and normalize


# Initialize the pipeline and specify the device
transcriber = pipeline(
    "automatic-speech-recognition",
    model="vinai/PhoWhisper-tiny",
    device=0 if torch.cuda.is_available() else -1,
)

input_path = scan_file_path_in_folder(path)
output_path = scan_file_name_in_folder(path)

for i in tqdm(range(0, len(input_path), 1)):
    audio_path = "Data/rawData/Audio/" + output_path[i] + ".mp3"
    # get the time of the audio file
    audio = AudioSegment.from_mp3(audio_path)
    duration = audio.duration_seconds

    # load csv file
    csv = pd.read_csv(input_path[i])

    # create folder "Data/PreporcessData/Audio_key_text/" + output_path[i] if not exist
    if not os.path.exists("Data/PreporcessData/Audio_key_text/" + output_path[i]):
        os.makedirs("Data/PreporcessData/Audio_key_text/" + output_path[i])
    for j in range(len(csv)):
        # check if txt file already exists then skip, fill j with 0 to 3 digits
        if os.path.exists(
            "Data/PreporcessData/Audio_key_text/"
            + output_path[i]
            + "/"
            + str(csv["n"][j]).zfill(3)
            + ".txt"
        ):
            continue
        # no grad
        with torch.no_grad():
            # get the text from the audio file
            # start is csv['pts_time'][j], end is csv['pts_time'][j+1] or the end of the audio file
            start = csv["pts_time"][j] * 1000  # pydub works in milliseconds
            if j == len(csv) - 1:
                end = duration * 1000
            else:
                end = csv["pts_time"][j + 1] * 1000
            audio_segment = audio[start:end]
            audio_np = audiosegment_to_numpy(audio_segment)

            text = transcriber(audio_np, return_timestamps=True)["text"]
        # write the text to the txt file
        with open(
            "Data/PreporcessData/Audio_key_text/"
            + output_path[i]
            + "/"
            + str(csv["n"][j]).zfill(3)
            + ".txt",
            "w",
            encoding="utf-8",
        ) as f:
            f.write(text)
```
